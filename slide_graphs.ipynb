```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
# Generate random data
n_observations = 100
n_classes = 3

# Generate random data for features and classes
X = np.random.rand(n_observations, 2) * 10  # Random X values between 0 and 10
y = np.random.randint(0, n_classes, n_observations)  # Random class labels from 0 to n_classes-1

# Create a new observation
new_observation = np.array([4.5, 2.5])


# Create a K-Nearest Neighbors classifier
k = 5
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X, y)

color_blue_light = "dodgerblue"
color_pink = "#e71d71"
color_gold = "#FFD700"

# Calculate the K-Nearest Neighbors
distances, indices = knn.kneighbors([new_observation])

# Get the labels of the K-Nearest Neighbors
neighbors_labels = y[indices]

# Create a color map with the three specified colors
colors = [color_blue_light, color_pink, color_gold]
cmap = plt.cm.colors.ListedColormap(colors)

# Plot the data with the nearest neighbors highlighted
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap)
plt.scatter(new_observation[0], new_observation[1], c='black', marker='o', s=100)
for neighbor in X[indices][0]:
    plt.plot([new_observation[0], neighbor[0]], [new_observation[1], neighbor[1]], 'k--', alpha=0.6)
plt.xlabel('Variabile X_1')
plt.ylabel('Variabile X_2')
plt.title('K-Nearest Neighbors con 3 classi')
plt.show()
```
